```python
from sklearn.model_selection import train_test_split

#Split the dataset into training and testing
X=d.iloc[:,2:33].values
Y=d.iloc[:,1].values
train_x,test_x,train_y,test_y=train_test_split(X,Y,test_size=0.30,random_state=1)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
SC=StandardScaler()
train_x=SC.fit_transform(train_x)
test_x=SC.fit_transform(test_x)
train_x
test_x
```

```python
#Function For fitting various Models
def ModelFit(train_x,train_y):
    #Logistic Regression
    from sklearn.linear_model import LogisticRegression
    log=LogisticRegression(random_state=0)
    log.fit(train_x,train_y)
    
    #Decision Tree
    from sklearn.tree import DecisionTreeClassifier
    clf_entropy=DecisionTreeClassifier(criterion="gini",random_state=0)
    clf_entropy.fit(train_x,train_y)
    
    #Random Forest
    from sklearn.ensemble import RandomForestClassifier
    Forest=RandomForestClassifier(n_estimators=100,criterion='entropy',random_state=0)
    Forest.fit(train_x,train_y)
    
    #SVM
    from sklearn import svm
    SVM=svm.SVC()
    SVM.fit(train_x,train_y)
    
    #KNN
    from sklearn.neighbors import KNeighborsClassifier
    clf=KNeighborsClassifier(n_neighbors=5)
    clf.fit(train_x,train_y)
    
    #Accuarcy on training data
    print('1.Logistic Regression Accuracy on training data',log.score(train_x,train_y))
    print('2.Decision Tree Accuracy on training data',clf_entropy.score(train_x,train_y))
    print('3.Random Forest Classifier Accuracy on training data',Forest.score(train_x,train_y))
    print('4.K Nearest Neighbour Accuracy on training data',Forest.score(train_x,train_y))
    print('5.Support Vector Machine Accuracy on training data',SVM.score(train_x,train_y))
    return log,clf_entropy,Forest,clf,SVM
```

```python
#Calling the function
Models=ModelFit(train_x,train_y)
```

```python
#Testing the accuracy using Confusion matrix
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
for i in range(len(Models)):
    print('Model:',i+1)
    CM=confusion_matrix(test_y,Models[i].predict(test_x))
    print(CM)
   
```

```python
for i in range(len(Models)):
    print('Model: ', i+1)
    print(classification_report(test_y,Models[i].predict(test_x)))
    Accuracy=accuracy_score(test_y,Models[i].predict(test_x))
    print('The accuracy on test data is:' ,Accuracy)
```
